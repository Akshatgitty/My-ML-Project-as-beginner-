```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv(r"C:\Users\aksha\OneDrive\Desktop\salary\salary_data.csv")
df
```

```python
x=df[["YearsExperience"]]
y=df[["Salary"]]
```

```python
x_train, x_test, y_train,y_test=train_test_split(x,y ,test_size=0.2,random_state=42)
model=LinearRegression()
```

```python

model.fit(x_train,y_train)
```

```python
user_input=float(input("Enter the Year of experience of have in work:"))
predicted_value=model.predict([[user_input]])
print(f"the salary you can expect is $ {predicted_value[0][0]:.2f}")
```

```python
plt.plot(x, model.predict(x), color='red')
plt.scatter(x,y)
plt.xlabel("years of experience")
plt.ylabel("Salary")
```

```python
model.score(x_test, y_test)
```
