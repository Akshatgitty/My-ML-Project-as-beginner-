```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression 
import matplotlib.pyplot as plt

df=pd.read_csv(r"C:\Users\aksha\OneDrive\Desktop\archive\score.csv")
df
```

```python
x=df[["Hours"]]
y=df[["Scores"]]

x_train,x_test, y_train, y_test=train_test_split(x,y, test_size=0.1 ,random_state=42)

model=LinearRegression()
```

```python
model.fit(x_train,y_train)

user_input=float(input("Enter the hours you study :"))

predicted_value=model.predict([[user_input]])

print(f"The predicted exam score you will have is {predicted_value[0][0]:.2f}")
```

```python
plt.plot(x, model.predict(x), color='red')
plt.scatter(x,y)
plt.xlabel("Hours Studied")
plt.ylabel("Scores")
plt.title("Hours vs Scores")

plt.show()
```

```python
plt.scatter(x_train, y_train, color='blue')
plt.scatter(x_test, y_test, color='green')

plt.plot(x_train, model.predict(x_train), color='red')
```

```python
acc=model.score(x_test,y_test)
print(" the accuracy this model is achieveing is " acc)
```
